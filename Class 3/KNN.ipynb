```python
a1 = np.array([5,7])
a2 = np.array([2,3])
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**0.5

def knn(X,Y,test_point,k=5):
    d = []
    m = X.shape[0]
    for i in range(m):
        dis = dist(test_point,X[i])
        d.append((dis,Y[i]))
        
    d.sort()
    d = np.array(d[:k])[:,1]
    uniq,freq = np.unique(d,return_counts=True)
    idx = np.argmax(freq)
    label = uniq[idx]
    
    return int(label)
```

```python
knn(X,Y,test_point)
```
