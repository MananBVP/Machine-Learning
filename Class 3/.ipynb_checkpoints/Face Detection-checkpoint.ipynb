```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./img1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img
```

```python
plt.imshow(img)
plt.axis("off")
```

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces = face_detector.detectMultiScale(img,1.3,5) #In 1.3 processing is fast but acc is low and in 1.1 processing is slow but acc is fast
```

```python
print(faces)
```

```python
for face in faces:
    x,y,w,h = face
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    
plt.figure(figsize=(15,15))
plt.imshow(img)
plt.axis("off")
plt.show()
```
